# Comparison of Leq values distribution in ARAUS original dataset and Fold-Fs, the generated testing fold

```python
import pandas as pd
import matplotlib.pyplot as plt
from matplotlib.ticker import FuncFormatter
plt.rcParams.update({
    'font.family': 'serif',
    'font.serif': ['Times New Roman'],  # Set to Times New Roman
    'font.size': 16,  # Set a suitable font size
    'axes.labelsize': 15,
    'axes.titlesize': 20,
    'legend.fontsize': 10,
    'xtick.labelsize': 13,
    'ytick.labelsize': 13
})
```

```python
# Paths to your CSV files
csv_file_path_1 = "../../../data/ARAUS_extended.csv"
csv_file_path_2 = "../../../data/fold_Fs.csv"

# Read CSV files into DataFrames
df1 = pd.read_csv(csv_file_path_1)
df2 = pd.read_csv(csv_file_path_2)

# Extract the 'info.Leq_R_r' columns
column1 = df1['info.Leq_R_r']
column2 = df2['info.Leq_R_r']

# Create histograms
plt.figure(figsize=[12,5],dpi=200)

# Define the formatter function
formatter = FuncFormatter(lambda y, _: f'{y:.2f}')

# Histogram for the first file
plt.subplot(1, 2, 1)
plt.gca().yaxis.set_major_formatter(formatter)
plt.hist(column1, density=True, bins=25, range=(65, 85),edgecolor='k', color='#0db2ac', alpha=0.7)
plt.title('Leq - ARAUS original dataset')
plt.xlabel('Leq values')

# Histogram for the second file
plt.subplot(1, 2, 2)
plt.gca().yaxis.set_major_formatter(formatter)
plt.hist(column2, density=True, bins=25, range=(65, 85),edgecolor='k', color='#fc694d',alpha=0.7)
plt.title('Leq - Fold Fs')
plt.xlabel('Leq values')



# Adjust layout and display the plots
plt.tight_layout()
plt.show()
```
