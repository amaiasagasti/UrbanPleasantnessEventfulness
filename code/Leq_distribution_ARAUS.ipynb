```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
df_ARAUS= pd.read_csv("../data/responses_SoundLights.csv")
print(df_ARAUS.columns)
```

```python
columns=['masker_bird','masker_silence','masker_construction','masker_traffic', 'masker_water', 'masker_wind','Leq_L_r', 'Leq_R_r']
df_ARAUS=df_ARAUS[columns]
print(df_ARAUS.columns)
```

```python
sources=['masker_bird','masker_construction', 'masker_traffic', 'masker_water', 'masker_wind']
for source in sources:
    mask = df_ARAUS[source] == 1
    data_ARAUS=df_ARAUS.loc[mask,'Leq_L_r']

    hist_values, bin_edges, _ =plt.hist(data_ARAUS, bins=200)  # Adjust the number of bins as needed
    plt.xlabel('Values')
    plt.ylabel('Frequency')
    title="Histogram Leq_r for "+source
    plt.title(title)
  

    # Find the maximum frequency and its corresponding bin edge
    max_freq_index = hist_values.argmax()
    max_freq = hist_values[max_freq_index]
    max_bin_edge = bin_edges[max_freq_index]

    print("Max frequency:", max_freq)
    print("Corresponding bin edge:", max_bin_edge)
    # Get the indices of the top 10 histogram values
    top_indices = np.argsort(hist_values)[::-1][:100]

    # Find the corresponding bin edges for these indices
    top_values = bin_edges[top_indices]

    print("Top 100 most common values: min", np.min(top_values), " and max ", np.max(top_values))

    plt.show()
```
