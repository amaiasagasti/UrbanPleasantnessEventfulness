```python
from essentia.standard import FreesoundExtractor
from Mosqito.loadFiles import load
import numpy as np
import pandas as pd
import os
import sys
sys.path.append("..")

print(os.listdir())

def calculate_stats(vector: np.array, stats: list, descriptor_name):
    output = {}
    """ if(np.ndim(vector)==1):
        along_axis=0
    elif(np.ndim(vector)==2):
        along_axis=1 """
    for i, stat in enumerate(stats):
        
        if stat == "avg":
            output[descriptor_name+"_avg"] = np.mean(vector, axis=0)
        if stat == "median":
            output[descriptor_name+"_median"] = np.median(vector, axis=0)
        if stat == "var":
            output[descriptor_name+"_var"] = np.var(vector, axis=0)
        """ if stat == "avgdB":
            output["avgdB"] = mean_dB(vector) """
        if stat == "rmc":
            output[descriptor_name+"_rmc"] = np.sqrt(np.mean(np.power(vector, 3, axis=0)))
        if stat == "max":
            output[descriptor_name+"_max"] = np.max(vector, axis=0)
        if stat == "min":
            output[descriptor_name+"_min"] = np.min(vector, axis=0)
        if stat == "p05":
            output[descriptor_name+"_p05"] = np.percentile(vector, 95, axis=0)
        if stat == "p10":
            output[descriptor_name+"_p10"] = np.percentile(vector, 90, axis=0)
        if stat == "p20":
            output[descriptor_name+"_p20"] = np.percentile(vector, 80, axis=0)
        if stat == "p30":
            output[descriptor_name+"_p30"] = np.percentile(vector, 70, axis=0)
        if stat == "p40":
            output[descriptor_name+"_p40"] = np.percentile(vector, 60, axis=0)
        if stat == "p50":
            output[descriptor_name+"_p50"] = np.percentile(vector, 50, axis=0)
        if stat == "p60":
            output[descriptor_name+"_p60"] = np.percentile(vector, 40, axis=0)
        if stat == "p70":
            output[descriptor_name+"_p70"] = np.percentile(vector, 30, axis=0)
        if stat == "p80":
            output[descriptor_name+"_p80"] = np.percentile(vector, 20, axis=0)
        if stat == "p90":
            output[descriptor_name+"_p90"] = np.percentile(vector, 10, axis=0)
        if stat == "p95":
            output[descriptor_name+"_p95"] = np.percentile(vector, 5, axis=0)
    return output



parameters = {
    'analysisSampleRate': 48000,
    'startTime': 0,
    'endTime': 30,

    'lowlevelFrameSize': 2048,
    'lowlevelHopSize': 1024,
    'lowlevelSilentFrames': 'noise',
    'lowlevelStats': ["mean"],#"max", "dmean", "dmean2", "dvar", "dvar2"
    'lowlevelWindowType': 'blackmanharris62',
    'lowlevelZeroPadding': 0,

    'mfccStats': ["mean"],
    "gfccStats":["mean"],

    'rhythmMaxTempo': 208,
    'rhythmMethod': 'degara',
    'rhythmMinTempo':40,
    'rhythmStats': ["mean"], #"mean", "var", "median", "min", "max", "dmean", "dmean2", "dvar", "dvar2"

    'tonalFrameSize': 4096,
    'tonalHopSize': 1024,
    'tonalSilentFrames': 'noise',
    'tonalStats': ["mean"],
    'tonalWindowType': 'blackmanharris62',
    'tonalZeroPadding': 0
}

def run_freesound_extractor(audiofile):
    """ Runs Essentia standard FreesoundExtractor
    :audiofile: absolute path to the audio file to analyze
    """
    try:
        result, resultFrames = FreesoundExtractor(**parameters)(audiofile)
    except RuntimeError as e:
        raise e
    return result, resultFrames

def analyze(input):
        # Run the essentia standard FreesoundExtractor
        #print('Running Essentia\'s FreesoundExtractor: {input}')
        result, frames = run_freesound_extractor(input)
        print(result)
        for descriptor in result.descriptorNames():
            print(descriptor, result[descriptor])
        
        # Calculate statistics and save all available features in a dictionary 
        features = dict()
        for descriptor in frames.descriptorNames():
            if(type(frames[descriptor])==np.ndarray):
                stats = [
                    "avg",
                    "median",
                    "var",
                    "min",
                    "max",
                    "p05",
                    "p10",
                    "p20",
                    "p30",
                    "p40",
                    "p50",
                    "p60",
                    "p70",
                    "p80",
                    "p90",
                    "p95",
                ]
                stats_result=calculate_stats(frames[descriptor], stats, descriptor)
                features = {**features, **stats_result}
            else:
                features[descriptor]=frames[descriptor]

        return features

new_df = pd.DataFrame(columns=feature_list)
# For each audio...
audios_path = "../data/soundscapes_augmented/"
for file in sorted(os.listdir(audios_path)):
    if file.endswith(".mp3") or file.endswith(".wav"):

        print("File ", file)
        audio_path = audios_path + file

        # Extract features for signal
        audio_features = analyze(audio_path)

        # Save
        new_df = pd.concat(
            [new_df, pd.DataFrame([audio_features])], ignore_index=True
        )  # Save row of data


new_df.to_csv("../data/generated_nonacoustic_features_SoundLights.csv", index=False)
```
